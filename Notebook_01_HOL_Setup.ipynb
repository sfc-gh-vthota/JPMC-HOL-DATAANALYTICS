## Step 1: Create Role

### Create the `NEXUS_ADMIN` role and grant it to ACCOUNTADMIN

````sql
%%sql -r dataframe_1
USE ROLE ACCOUNTADMIN;

-- Create lab role
CREATE ROLE IF NOT EXISTS NEXUS_ADMIN
  COMMENT = 'Summit 2026 HOL - Nexus Capital lab role';

-- Grant role to current user and ACCOUNTADMIN
GRANT ROLE NEXUS_ADMIN TO ROLE ACCOUNTADMIN;

SELECT 'NEXUS_ADMIN role created and granted' AS STATUS;
````

## Step 2: Create Database & Schemas

### Create `NEXUS_HOL` with 4 schemas:
- **ANALYTICS** — Gold-layer analytical tables (pre-loaded)
- **SEMANTIC** — Semantic views and governed metrics
- **AGENTS** — Cortex Agents, SI objects, Cortex Search services, MCP servers
- **TRANSACTIONS** — Reserved for optional Postgres streaming module

````sql
%%sql -r dataframe_2
-- Create database
CREATE DATABASE IF NOT EXISTS NEXUS_HOL
  COMMENT = 'Summit 2026 HOL - Nexus Capital AI Enterprise Application';

-- Create schemas
CREATE SCHEMA IF NOT EXISTS NEXUS_HOL.ANALYTICS
  COMMENT = 'Gold-layer analytical tables for Nexus Capital';

CREATE SCHEMA IF NOT EXISTS NEXUS_HOL.SEMANTIC
  COMMENT = 'Semantic views and governed business metrics';

CREATE SCHEMA IF NOT EXISTS NEXUS_HOL.AGENTS
  COMMENT = 'Cortex Agents, SI objects, Cortex Search services, MCP servers';

CREATE SCHEMA IF NOT EXISTS NEXUS_HOL.TRANSACTIONS
  COMMENT = 'Reserved for optional Postgres streaming module';

-- Grant ownership to NEXUS_ADMIN
GRANT OWNERSHIP ON DATABASE NEXUS_HOL TO ROLE NEXUS_ADMIN COPY CURRENT GRANTS;
GRANT OWNERSHIP ON ALL SCHEMAS IN DATABASE NEXUS_HOL TO ROLE NEXUS_ADMIN COPY CURRENT GRANTS;

SELECT 'Database NEXUS_HOL created with 4 schemas' AS STATUS;
````

## Step 3: Create Warehouse

### Create `NEXUS_WH` — XS size with auto-suspend for cost efficiency

````sql
%%sql -r dataframe_3
-- Create warehouse
CREATE WAREHOUSE IF NOT EXISTS NEXUS_WH
  WAREHOUSE_SIZE = 'XSMALL'
  AUTO_SUSPEND = 60
  AUTO_RESUME = TRUE
  INITIALLY_SUSPENDED = TRUE
  COMMENT = 'Summit 2026 HOL - Nexus Capital compute';

-- Grant usage to NEXUS_ADMIN
GRANT USAGE ON WAREHOUSE NEXUS_WH TO ROLE NEXUS_ADMIN;
GRANT OPERATE ON WAREHOUSE NEXUS_WH TO ROLE NEXUS_ADMIN;

SELECT 'NEXUS_WH warehouse created (XS, auto-suspend 60s)' AS STATUS;
````

## Step 4: Grant Privileges

### Grant all required privileges for:
- CoWork (formerly Snowflake Intelligence) (Agents, SI Objects)
- Cortex services (Search, Analyst, Semantic Views)
- MCP Servers (External and Managed)
- Optional: Postgres instance creation, Dynamic Tables

````sql
%%sql -r dataframe_4
-- Account-level grants for SI, Agents, and Cortex
GRANT CREATE SNOWFLAKE INTELLIGENCE ON ACCOUNT TO ROLE NEXUS_ADMIN;
GRANT CREATE INTEGRATION ON ACCOUNT TO ROLE NEXUS_ADMIN;
GRANT EXECUTE MANAGED TASK ON ACCOUNT TO ROLE NEXUS_ADMIN;

-- Cortex Agent and database roles
GRANT DATABASE ROLE SNOWFLAKE.CORTEX_USER TO ROLE NEXUS_ADMIN;

-- Schema-level grants: SEMANTIC schema
GRANT CREATE SEMANTIC VIEW ON SCHEMA NEXUS_HOL.SEMANTIC TO ROLE NEXUS_ADMIN;

-- Schema-level grants: AGENTS schema
GRANT CREATE CORTEX SEARCH SERVICE ON SCHEMA NEXUS_HOL.AGENTS TO ROLE NEXUS_ADMIN;
GRANT CREATE AGENT ON SCHEMA NEXUS_HOL.AGENTS TO ROLE NEXUS_ADMIN;
GRANT CREATE MCP SERVER ON SCHEMA NEXUS_HOL.AGENTS TO ROLE NEXUS_ADMIN;
GRANT CREATE EXTERNAL MCP SERVER ON SCHEMA NEXUS_HOL.AGENTS TO ROLE NEXUS_ADMIN;
GRANT CREATE NETWORK RULE ON SCHEMA NEXUS_HOL.AGENTS TO ROLE NEXUS_ADMIN;

-- Schema-level grants: ANALYTICS schema
GRANT CREATE TABLE ON SCHEMA NEXUS_HOL.ANALYTICS TO ROLE NEXUS_ADMIN;
GRANT CREATE VIEW ON SCHEMA NEXUS_HOL.ANALYTICS TO ROLE NEXUS_ADMIN;
GRANT CREATE DYNAMIC TABLE ON SCHEMA NEXUS_HOL.ANALYTICS TO ROLE NEXUS_ADMIN;
GRANT CREATE STREAM ON SCHEMA NEXUS_HOL.ANALYTICS TO ROLE NEXUS_ADMIN;
GRANT CREATE TASK ON SCHEMA NEXUS_HOL.ANALYTICS TO ROLE NEXUS_ADMIN;

-- Schema-level grants: TRANSACTIONS schema (for optional Postgres module)
GRANT CREATE TABLE ON SCHEMA NEXUS_HOL.TRANSACTIONS TO ROLE NEXUS_ADMIN;
GRANT CREATE STAGE ON SCHEMA NEXUS_HOL.TRANSACTIONS TO ROLE NEXUS_ADMIN;

-- Optional: Postgres instance creation (for Module 10)
GRANT CREATE POSTGRES INSTANCE ON ACCOUNT TO ROLE NEXUS_ADMIN;

SELECT 'All privileges granted to NEXUS_ADMIN' AS STATUS;
````

## Step 6: Switch to NEXUS_ADMIN and Load Gold-Layer Data

### Now we populate the ANALYTICS schema with Nexus Capital's gold-layer tables:
- **CLIENTS** — Client master data (institutional, HNW accounts)
- **POSITIONS** — Current portfolio positions by client and symbol
- **TRADES** — Recent trade executions
- **DAILY_METRICS** — Aggregated daily business metrics (AUM, revenue, risk)

This represents the analytical layer that would typically be fed by streaming/CDC from transactional systems. For this lab, it's pre-loaded so you can jump straight into building the semantic layer and agents.

````sql
%%sql -r dataframe_6
USE ROLE NEXUS_ADMIN;
USE DATABASE NEXUS_HOL;
USE SCHEMA ANALYTICS;
USE WAREHOUSE NEXUS_WH;
````

````sql
%%sql -r dataframe_7
-- CLIENTS table: Client master data
CREATE OR REPLACE TABLE NEXUS_HOL.ANALYTICS.CLIENTS (
    CLIENT_ID           NUMBER AUTOINCREMENT PRIMARY KEY,
    CLIENT_NAME         VARCHAR(100) NOT NULL,
    ACCOUNT_TYPE        VARCHAR(20) NOT NULL,      -- INSTITUTIONAL, RETAIL, HNW
    REGION              VARCHAR(30) NOT NULL,
    AUM                 NUMBER(15,2),              -- Assets Under Management
    RISK_PROFILE        VARCHAR(20),               -- CONSERVATIVE, MODERATE, AGGRESSIVE
    ONBOARDED_DATE      DATE NOT NULL,
    RELATIONSHIP_MANAGER VARCHAR(50),
    EMAIL               VARCHAR(100),
    STATUS              VARCHAR(20) DEFAULT 'ACTIVE'
);

-- POSITIONS table: Current portfolio positions
CREATE OR REPLACE TABLE NEXUS_HOL.ANALYTICS.POSITIONS (
    POSITION_ID         NUMBER AUTOINCREMENT PRIMARY KEY,
    CLIENT_ID           NUMBER NOT NULL,
    SYMBOL              VARCHAR(10) NOT NULL,
    QUANTITY            NUMBER NOT NULL,
    AVG_COST            NUMBER(12,4) NOT NULL,
    CURRENT_PRICE       NUMBER(12,4),
    MARKET_VALUE        NUMBER(15,2),
    UNREALIZED_PNL      NUMBER(15,2),
    SECTOR              VARCHAR(30),
    LAST_UPDATED        TIMESTAMP_NTZ DEFAULT CURRENT_TIMESTAMP()
);

-- TRADES table: Recent trade executions
CREATE OR REPLACE TABLE NEXUS_HOL.ANALYTICS.TRADES (
    TRADE_ID            NUMBER AUTOINCREMENT PRIMARY KEY,
    CLIENT_ID           NUMBER NOT NULL,
    SYMBOL              VARCHAR(10) NOT NULL,
    TRADE_TYPE          VARCHAR(4) NOT NULL,       -- BUY or SELL
    QUANTITY            NUMBER NOT NULL,
    PRICE               NUMBER(12,4) NOT NULL,
    TRADE_DATE          TIMESTAMP_NTZ NOT NULL DEFAULT CURRENT_TIMESTAMP(),
    SETTLEMENT_DATE     DATE,
    STATUS              VARCHAR(20) DEFAULT 'EXECUTED',
    EXCHANGE            VARCHAR(10) DEFAULT 'NYSE',
    NOTES               VARCHAR(500)
);

-- DAILY_METRICS table: Aggregated daily business metrics
CREATE OR REPLACE TABLE NEXUS_HOL.ANALYTICS.DAILY_METRICS (
    METRIC_DATE         DATE NOT NULL,
    TOTAL_AUM           NUMBER(18,2),
    TOTAL_CLIENTS       NUMBER,
    ACTIVE_CLIENTS      NUMBER,
    TOTAL_TRADES        NUMBER,
    TRADE_VOLUME_USD    NUMBER(18,2),
    NET_FLOWS           NUMBER(15,2),
    TOTAL_REVENUE       NUMBER(15,2),
    AVG_PORTFOLIO_RISK  NUMBER(5,3),
    TOP_SECTOR          VARCHAR(30)
);

SELECT 'Analytics tables created' AS STATUS;
````
